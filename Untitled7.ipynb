```python
import numpy as np
import tensorflow as tf
from tensorflow import keras as K
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(11)
tf.random.set_seed(11)

# Hyperparameters
batch_size = 256
max_epochs = 50
learning_rate = 1e-3

hidden_dim = 128
latent_dim = 32
original_dim = 784

# Load MNIST dataset
(x_train, _), (x_test, _) = K.datasets.mnist.load_data()

print("Training data shape before reshaping:", x_train.shape)
print("Testing data shape before reshaping:", x_test.shape)

# Normalize pixel values from 0-255 to 0-1
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# Flatten each 28x28 image into 784 values
x_train = np.reshape(x_train, (x_train.shape[0], original_dim))
x_test = np.reshape(x_test, (x_test.shape[0], original_dim))

print("Training data shape after reshaping:", x_train.shape)
print("Testing data shape after reshaping:", x_test.shape)

# Create training dataset
training_dataset = tf.data.Dataset.from_tensor_slices(x_train)
training_dataset = training_dataset.shuffle(buffer_size=1024).batch(batch_size)
```

```python
class Encoder(K.layers.Layer):
    def __init__(self, hidden_dim, latent_dim):
        super(Encoder, self).__init__()

        self.hidden_layer = K.layers.Dense(
            units=hidden_dim,
            activation=tf.nn.relu
        )

        self.latent_layer = K.layers.Dense(
            units=latent_dim,
            activation=tf.nn.relu
        )

    def call(self, input_features):
        hidden_activation = self.hidden_layer(input_features)
        latent_output = self.latent_layer(hidden_activation)
        return latent_output
```

```python
class Decoder(K.layers.Layer):
    def __init__(self, hidden_dim, original_dim):
        super(Decoder, self).__init__()

        self.hidden_layer = K.layers.Dense(
            units=hidden_dim,
            activation=tf.nn.relu
        )

        self.output_layer = K.layers.Dense(
            units=original_dim,
            activation=tf.nn.sigmoid
        )

    def call(self, encoded):
        hidden_activation = self.hidden_layer(encoded)
        reconstruction = self.output_layer(hidden_activation)
        return reconstruction
```

```python
class Autoencoder(K.Model):
    def __init__(self, hidden_dim, latent_dim, original_dim):
        super(Autoencoder, self).__init__()

        self.encoder = Encoder(
            hidden_dim=hidden_dim,
            latent_dim=latent_dim
        )

        self.decoder = Decoder(
            hidden_dim=hidden_dim,
            original_dim=original_dim
        )

    def call(self, input_features):
        encoded = self.encoder(input_features)
        reconstructed = self.decoder(encoded)
        return reconstructed
```

```python
autoencoder = Autoencoder(
    hidden_dim=hidden_dim,
    latent_dim=latent_dim,
    original_dim=original_dim
)

optimizer = K.optimizers.Adam(learning_rate=learning_rate)

def loss(model, original):
    reconstruction = model(original)

    reconstruction_error = tf.reduce_mean(
        tf.square(tf.subtract(reconstruction, original))
    )

    return reconstruction_error
```

```python
# Build model before summary
autoencoder.build(input_shape=(None, original_dim))

# Display model summary
autoencoder.summary()
```

```python
def train(loss, model, optimizer, original):
    with tf.GradientTape() as tape:
        reconstruction_error = loss(model, original)

    gradients = tape.gradient(
        reconstruction_error,
        model.trainable_variables
    )

    optimizer.apply_gradients(
        zip(gradients, model.trainable_variables)
    )

    return reconstruction_error
```

```python
def train_loop(model, optimizer, loss, dataset, epochs):
    model.losses_list = []

    for epoch in range(epochs):
        epoch_loss = 0

        for step, batch_features in enumerate(dataset):
            loss_value = train(loss, model, optimizer, batch_features)
            epoch_loss += loss_value

        model.losses_list.append(epoch_loss.numpy())

        print(
            "Epoch {}/{}. Loss: {}".format(
                epoch + 1,
                epochs,
                epoch_loss.numpy()
            )
        )


train_loop(
    autoencoder,
    optimizer,
    loss,
    training_dataset,
    epochs=max_epochs
)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, max_epochs + 1), autoencoder.losses_list)

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Autoencoder Training Loss over 50 Epochs")

plt.savefig("Figure_1_Autoencoder_Training_Loss.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
number = 10
plt.figure(figsize=(20, 4))

for index in range(number):
    # Original image
    ax = plt.subplot(2, number, index + 1)
    plt.imshow(x_test[index].reshape(28, 28), cmap="gray")
    plt.title("Original")
    plt.axis("off")

    # Reconstructed image
    reconstruction = autoencoder(tf.constant(x_test[index:index + 1]))

    ax = plt.subplot(2, number, index + 1 + number)
    plt.imshow(tf.reshape(reconstruction, (28, 28)), cmap="gray")
    plt.title("Reconstructed")
    plt.axis("off")

plt.suptitle("Original and Reconstructed MNIST Digits")
plt.savefig("Figure_2_Original_vs_Reconstructed.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
original_dimension = 784
compressed_dimension = latent_dim

reduction_ratio = compressed_dimension / original_dimension
remaining_percentage = reduction_ratio * 100
compression_percentage = (1 - reduction_ratio) * 100

print("Original dimension:", original_dimension)
print("Compressed dimension:", compressed_dimension)
print("Reduction ratio:", reduction_ratio)
print("Remaining percentage:", remaining_percentage)
print("Compression percentage:", compression_percentage)
```
